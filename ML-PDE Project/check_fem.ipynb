```python
nx = 32
ny = 32

# [bottom, top, left, right]
bc_values = [[0,0,0,0],[None,None,0,.5],[0,.5,None,None],[1,None,None,1]]

for bc_value in bc_values:
    # input image
    img = np.zeros((ny,nx)) - 2
    if bc_value[0] is not None: img[0, :] = bc_value[0]  #bottom
    if bc_value[1] is not None: img[-1,:] = bc_value[1]  #top
    if bc_value[2] is not None: img[:, 0] = bc_value[2]  #left
    if bc_value[3] is not None: img[:,-1] = bc_value[3]  #right

    # Solving with FEM
    U = fem.solve(img,2)

    # Plotting the reuslts
    uts.plot_results(img,U)
```
